# Task 1: Matplotlib Basics

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
plt.plot(x, y, label='Line Plot')
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.savefig('line_plot.png', dpi=300)
plt.show()

# Scatter plot
plt.scatter(x, y, color='red', label='Scatter Plot')
plt.title('Basic Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.savefig('scatter_plot.png', dpi=300)
plt.show()
```

# Task 2: Plot Customization

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='green', linestyle='--', marker='o', label='Customized Line')
plt.title('Customized Plot')
plt.xlabel('Custom X')
plt.ylabel('Custom Y')
plt.legend()
plt.grid(True)
plt.xlim(0, 6)
plt.ylim(5, 35)
plt.savefig('custom_plot.png', dpi=300)
plt.show()
```

# Task 3: Seaborn Introduction

```python
import seaborn as sns
import numpy as np
import pandas as pd
data = pd.DataFrame({
    'A': np.random.normal(0, 1, 100),
    'B': np.random.normal(5, 2, 100),
    'C': np.random.randint(1, 5, 100)
})

# Boxplot
sns.boxplot(x='C', y='B', data=data)
plt.title('Boxplot Example')
plt.savefig('boxplot.png', dpi=300)
plt.show()

# Heatmap (correlation matrix)
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation')
plt.savefig('heatmap.png', dpi=300)
plt.show()

# Histogram
sns.histplot(data['A'], kde=True)
plt.title('Histogram of A')
plt.savefig('histogram.png', dpi=300)
plt.show()
```

# Task 4: Sample Visualization Project

```python
tips = sns.load_dataset("tips")
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title('Total Bill Distribution')
plt.savefig('tips_histogram.png', dpi=300)
plt.show()

# Line plot (average total bill per day)
avg_tips = tips.groupby('day', observed=True)['total_bill'].mean()

plt.plot(avg_tips.index, avg_tips.values, marker='o')
plt.title('Average Total Bill per Day')
plt.xlabel('Day')
plt.ylabel('Avg Total Bill')
plt.savefig('tips_lineplot.png', dpi=300)
plt.show()

# Boxplot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Total Bill by Day')
plt.savefig('tips_boxplot.png', dpi=300)
plt.show()

```

# Task 5: Produce Publish-Quality Figures

```python
plt.figure(figsize=(10, 6), dpi=300)
sns.boxplot(x='day', y='total_bill', hue='day', data=tips, palette='pastel',legend=False)
plt.title('Total Bill by Day (High Quality)', fontsize=14)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Total Bill (USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('publish_quality_figure.png', dpi=300)
plt.show()
```
