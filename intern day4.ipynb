# Task 1: Indexing and Slicing



```python
import pandas as pd
df = pd.read_csv("sales_data_with_discounts.csv")
df.head()
```

```python
print("Model at row label 0:", df.loc[0, 'Model'])
print("Brand at row 2, column 5:", df.iloc[2, 5])
print("Entire row using loc:")
print(df.loc[3])
print("\nEntire row using iloc:")
print(df.iloc[3])
df[['Date', 'Brand', 'Net Sales Value']].head()
high_value_sales = df[df['Net Sales Value'] > 100000]
print("Sales with Net Sales Value > 100000:")
print(high_value_sales[['Model', 'Net Sales Value']])
high_discount_sales = df[df['Discount Rate (%)'] > 10]
print("\nProducts with Discount Rate > 10%:")
print(high_discount_sales[['Brand', 'Discount Rate (%)']])
```
